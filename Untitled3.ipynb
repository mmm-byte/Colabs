```r
missing_values <- colSums(is.na(data))
print(missing_values)
```

```r
library(ggplot2)
missing_vals <- sapply(data, function(x) sum(is.na(x)))
missing_df <- data.frame(
  column = names(missing_vals),
  count = missing_vals
)
ggplot(missing_df, aes(x = column, y = count)) +
  geom_bar(stat = "identity") +
  ylab("% Missing") +
  xlab("Variables") +
  ggtitle("Missing values per column")+coord_flip()
```
